**1. Importación de Librerias y Configuración**

```python
# Importación de librerías esenciales
import pandas as pd      # Para la manipulación y análisis de datos
import numpy as np       # Para operaciones numéricas
import matplotlib.pyplot as plt  # Para la generación de gráficos
import seaborn as sns    # Para visualización de datos con estilos predefinidos

# Configuración de estilo para gráficos
sns.set(style="whitegrid", palette="muted", color_codes=True)
plt.rcParams['figure.figsize'] = (10, 6)  # Tamaño predeterminado de las figuras

# Configuración de opciones de Pandas
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
```

**2. Carga de Datos y Exploración Inicial**

```python
# Cargar el dataset variables utilizando Pandas
data_energia = pd.read_excel('(1) Energy 2000-2015 .xlsx')

# Mostrar las primeras filas de cada DataFrame para confirmar la carga
print("Primeras filas del dataset de Energías:")
display(data_energia.head())


#Mostrar las ultimas filas de cada DataFrame para confirmar la carga
print("Ultimas filas del dataset de Energías:")
display(data_energia.tail())


# información general (tipos de datos, cantidad de registros, etc.)
print("Información del dataset de Energías:")
print(data_energia.info())



#verificación de los tipos de datos
print("Tipos de datos del dataset de Energías:")
print(data_energia.dtypes)


#verificación de valores nulos
print("Valores nulos del dataset de Energías:")
print(data_energia.isnull().sum())


#verificación de valores duplicados
print("Valores duplicados del dataset de Energías:")
print(data_energia.duplicated().sum())
```

**3. Limpieza y preparación de los datos**

```python

"""Para cada una de las siguientes variables se calcula para cada país la media y se reemplaza en los valores nulos (si tiene en ese país)

# Columnas con valores nulos

# # 1. Access to electricity, rural (% of rural population)
# # 2. Combustible renewables and waste (% of total energy)
# # 3. Electricity production from nuclear sources (% of total)
# # 4. Electric power consumption (kWh per capita)
# # 5. Fossil fuel energy consumption (% of total)
# # 6. GDP per capita (constant LCU)
# # 7. GDP per unit of energy use (constant 2021 PPP $ per kg of oil equivalent)
# # 8. GDP (constant LCU)
"""

# Para Access to electricity, rural (% of rural population)
# Calcular la media por país y reemplazar los valores nulos
data_energia['Access to electricity, rural (% of rural population)'] = data_energia.groupby('Country Name')['Access to electricity, rural (% of rural population)'].transform(lambda x: x.fillna(x.mean()))

# Para Combustible renewables and waste (% of total energy)
data_energia['Combustible renewables and waste (% of total energy)'] = data_energia.groupby('Country Name')['Combustible renewables and waste (% of total energy)'].transform(lambda x: x.fillna(x.mean()))

# Para Electricity production from nuclear sources (% of total)
data_energia['Electricity production from nuclear sources (% of total)'] = data_energia.groupby('Country Name')['Electricity production from nuclear sources (% of total)'].transform(lambda x: x.fillna(x.mean()))

# Para Electric power consumption (kWh per capita)
data_energia['Electric power consumption (kWh per capita)'] = data_energia.groupby('Country Name')['Electric power consumption (kWh per capita)'].transform(lambda x: x.fillna(x.mean()))

# Fossil fuel energy consumption (% of total)
data_energia['Fossil fuel energy consumption (% of total)'] = data_energia.groupby('Country Name')['Fossil fuel energy consumption (% of total)'].transform(lambda x: x.fillna(x.mean()))
# Para GDP per capita (constant LCU)
data_energia['GDP per capita (constant LCU)'] = data_energia.groupby('Country Name')['GDP per capita (constant LCU)'].transform(lambda x: x.fillna(x.mean()))

# Para GDP per unit of energy use (constant 2021 PPP $ per kg of oil equivalent)
data_energia['GDP per unit of energy use (constant 2021 PPP $ per kg of oil equivalent)'] = data_energia.groupby('Country Name')['GDP per unit of energy use (constant 2021 PPP $ per kg of oil equivalent)'].transform

# Para GDP (constant LCU)
data_energia['GDP (constant LCU)'] = data_energia.groupby('Country Name')['GDP (constant LCU)'].transform(lambda x: x.fillna(x.mean()))

print(data_energia.isnull().sum()) # Verificamos que no hayan valores nulos
```

# **`3. Comparación entre la producción eléctrica y las emisiones de gases de efecto invernadero.`**


Aquí se calcula la intensidad de carbono de la producción de energía.
Se divide las emisiones totales de GEI (excluyendo LULUCF) por la producción total de energía (kWh. Como resultado obtenemos cuántas Mega toneladas de CO2e se emiten por cada kilovatio-hora de electricidad producido.

```python
font1 = {'family':'serif','color':'black','fontsize':15} # Podria ir size pero para que funcione para plt.suptitle debe ser fontsize
font2 = {'family':'serif','color':'black','fontsize':13}
font3 = {'family':'serif','size':12} # Para la leyenda (plt.legend)
color_names = ["Aqua","blue", "blueviolet","yellow", "darkviolet", "chartreuse", "coral","chocolate","crimson", "darkcyan","dimgray","darkgreen", "deeppink", "darkgoldenrod", "fuchsia", "gold","indigo","lawngreen","lightsalmon", "lime", "magenta", "maroon", "mediumspringgreen", "moccasin", "olive", "Orange", "peru", "red", "brown","rebeccapurple","tan", "mistyrose", "navy","black", "lightcyan"]
# 1. Calcular la intensidad de carbono

data_energia['Intensidad de Carbono'] = data_energia['Total greenhouse gas emissions excluding LULUCF (Mt CO2e)'] / (data_energia['Total electricity output (GWh) [4.1.1_TOTAL.ELECTRICITY.OUTPUT]'] *1000000)  # Multiplicar por 1000000 para convertir GWh a kWh


# 2. Filtrar datos para Colombia
data_colombia = data_energia[data_energia['Country Name'] == 'Colombia']


# 3. Crear gráficos de líneas

# Gráfico para todos los países
plt.figure(figsize=(15, 8))
sns.lineplot(data=data_energia, x='Time', y='Intensidad de Carbono', hue='Country Name', palette=color_names)
plt.title('Emisiones de CO2e por unidad de electricidad producida (2000-2015)', fontdict=font1)
plt.xlabel('Año', fontdict=font2)
plt.ylabel('Toneladas de CO2e por kWh', fontdict=font2)
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left', prop=font3)
plt.show()


# Gráfico para Colombia
plt.figure(figsize=(10, 6))
sns.lineplot(data=data_colombia, x='Time', y='Intensidad de Carbono', color = 'coral')
plt.title('Emisiones de CO2e por unidad de electricidad producida (Colombia)(2000-2015)', fontdict=font1)
plt.xlabel('Año', fontdict=font2)
plt.ylabel('Mega toneladas de CO2e por kWh', fontdict=font2)
plt.show()
```
